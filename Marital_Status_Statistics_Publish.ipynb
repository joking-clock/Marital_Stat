```python
# labels = ['G1', 'G2', 'G3', 'G4', 'G5']
# men_means = [20, 34, 30, 35, 27]
# women_means = [25, 32, 34, 20, 25]
plt.rcParams["figure.figsize"] = (20,3)


labels = men_usa[men_usa['YearStart']=='1975-01-01'].loc[:,'AgeGroup'].tolist()
men = men_usa[men_usa['YearStart']=='1975-01-01'].loc[:,'DataValue'].values
women = women_usa[women_usa['YearStart']=='1975-01-01'].loc[:,'DataValue'].values

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men, width, label='Men')
rects2 = ax.bar(x + width/2, women, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Married percentages')
ax.set_title('Married percentages of USA in 1975')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

plt.ylim(0,120)

plt.show()
```
